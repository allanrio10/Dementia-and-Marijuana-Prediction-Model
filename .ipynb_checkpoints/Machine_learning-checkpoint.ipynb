```python
#Filling missing Values 
#create function automating missing value in this dataset
def fill_nulls(data):
    '''
    This functions explores the dataframe, looks for missing values, and fills missing values with zero
    then returns missing values for each column in the data
    '''
    missing_values = data.isna().sum()
    
    for col in data.columns:
        if missing_values[col]> 0:
            data[col].fillna(0, inplace= True)
            missing_values[col] = 0 
            
    return missing_values
```

```python
#check for any special characters
def special_characters_checker(data):
    '''
    Functions check for special characters in the data and returns the count
    '''
    count_characters = data.apply(lambda x: x.astype(str).str.count(r'[^a-zA-Z0-9\s]').sum())
    return count_characters
special_counts = special_characters_checker(df)
print("Count of Special Characters:")
print(special_counts)
```

```python
#Columns suggested having special characters
columns = df[["age1bir", "bmi", "drunk", "usedpot", "usedcrck", "PhysicalHlth", "MentalHlth", "MEIS"]]

#check for special characters in these columns

def special_characters_checker(columns):
    '''
    Function to check for special characters in each specified column of a DataFrame.
    '''
    special_characters_dict = {}
    special_characters = set(string.punctuation)  # set of special characters
    
    for column_name in columns.columns:
        column_values = columns[column_name].astype(str)
        special_chars_found = set(''.join(column_values.tolist())) & special_characters
        special_characters_dict[column_name] = special_chars_found
    
    return special_characters_dict

# Call the function
special_chars_found = special_characters_checker(columns)

# Print special characters found for each column
for column_name, chars_found in special_chars_found.items():
    print(f"Column: {column_name}")
    print(chars_found)
    print("---------------------")
```

```python
#Check for Unique values in the given columns 
# specified columns of interest
columns = df[["age1bir", "bmi", "drunk", "usedpot", "usedcrck", "PhysicalHlth", "MentalHlth", "MEIS"]]

def check_unique_values(columns):
    '''
    Function to check unique values in each specified column of a DataFrame.
    '''
    unique_values_dict = {}
    
    for column_name in columns.columns:
        unique_values = columns[column_name].unique()
        unique_values_dict[column_name] = unique_values
    
    return unique_values_dict

# Call the function
unique_values = check_unique_values(columns)

# Print unique values for each column
for column_name, values in unique_values.items():
    print(f"Column: {column_name}")
    print(values)
    print("---------------------")
```

###### Conclusion   
Keep the special characters for accurate stastical representations

```python
#Isolate dependent and independent variables 
X = df.drop(columns = "MEIS")
y = df["MEIS"]
```

c. Standardizing scale    

```python
X.head(3).iloc[:,:9]
```

```python
X.info()
```

```python
# Separate categorical columns
ordinal_columns = ['drunk', 'usedpot', 'usedcrck', 'FeltDown', 'marital']  # categorical ordinal columns
nominal_columns = ['smoker', 'poverty', 'educatn', 'racethn']  # categorical nominal columns

# Instantiate LabelEncoder and convert ordinal columns
label_encoder = LabelEncoder()
X_encoded = X.copy()  # Create a copy of X to work with

for col in ordinal_columns:
    X_encoded[col] = label_encoder.fit_transform(X[col])

# Instantiate OneHotEncoder and convert nominal columns
ohe = OneHotEncoder(sparse_output = False, handle_unknown='ignore')

# Transform nominal data correctly
nominal_data = X[nominal_columns].values
nominal_data = ohe.fit_transform(nominal_data)

# Create DataFrame for nominal columns with appropriate column names
nominal_df = pd.DataFrame(nominal_data, columns=ohe.get_feature_names_out(nominal_columns))

# Drop original nominal columns from X_encoded
X_encoded.drop(columns=nominal_columns, inplace=True)

# Concatenate nominal_df and ordinal encoded columns with original data
X_encoded = pd.concat([X_encoded, nominal_df], axis=1)

# Display the encoded DataFrame
X_encoded.head(4).iloc[:,:]
```

```python
#Transform scale but first review the datatypes 
X_encoded.dtypes
```

```python
#convert docvist to numerical
# Replace 'DK' with NaN
X_encoded["docvisit"] = pd.to_numeric(X_encoded["docvisit"], errors='coerce')

# Convert to int64
X_encoded["docvisit"] = X_encoded["docvisit"].astype('Int64')

# Convert int32 columns to int64
int32_columns = X_encoded.select_dtypes(include='int32').columns
X_encoded[int32_columns] = X_encoded[int32_columns].astype('int64')

# Display updated DataFrame i
X_encoded.info()
```

```python
#Convert to uniform scaling using MinMax Scaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_encoded)

#convert scaled dataframe into original dataframe 
X_encoded = pd.DataFrame(X_scaled, columns = X_encoded.columns)

#preview scale
X_encoded.describe()
```

```python
#recheck for possible missing values 
X_encoded.isna().sum()
```

```python
#call out function to fill missing values
fill_nulls(X_encoded)
```

```python
#split encoded data into train and test
X_train, X_test, y_train, y_test = train_test_split(X_encoded,y, test_size = 0.25, random_state =42)
```

```python
#instantiate model
model = LinearRegression()

#fit training data
model.fit(X_train,y_train)
```

```python
# Evaluate Performance of Model
train_pred = model.predict(X_train)
train_mse = mean_squared_error(y_train, train_pred)
print(f"Training MSE: {train_mse}")

# Evaluate on test set
y_pred = model.predict(X_test)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE: {test_mse}")
```

```python
#Visualize Model Performance 
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, color='blue', alpha=0.5)
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'k--', lw=2)
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs Predicted')
plt.show()
```

```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error
import numpy as np

# Assuming you have already defined Important_features and y

# Check for shape
print("Important_features shape:", Important_features.shape)
print("Shape of y:", y.shape)

# Split encoded data into train and test
X_train, X_test, y_train, y_test = train_test_split(Important_features, y, test_size=0.25, random_state=42)

# Scale the data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Instantiate model with StratifiedKFold and increased max_iter
model = LogisticRegressionCV(cv=StratifiedKFold(n_splits=2), max_iter=1000)  # Adjust max_iter as needed

# Fit training data
model.fit(X_train_scaled, y_train)

# Evaluate Performance of Model
train_pred = model.predict(X_train_scaled)
train_mse = mean_squared_error(y_train, train_pred)
print(f"Training MSE: {train_mse}")

# Evaluate on test set
y_pred = model.predict(X_test_scaled)
test_mse = mean_squared_error(y_test, y_pred)
print(f"Test MSE: {test_mse}")
```
